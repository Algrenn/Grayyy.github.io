```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv("stock.csv")


data = {
    'Purchase Price': [150, 200, 180, 300, 250, 280, 210, 190, 220, 240, 260, 270, 230, 320, 340],
    'Current Price': [160, 180, 170, 290, 260, 290, 200, 180, 210, 230, 250, 270, 220, 310, 330],
    'Gain/Loss': [10, -20, -10, -10, 10, 10, -10, -10, -10, -10, -10, 0, -10, -10, -10]
}
df = pd.DataFrame(data)

plt.figure(figsize=(12, 8))


plt.subplot(2, 2, 1)
plt.hist(df['Purchase Price'], bins=15, color='#9C27B0', alpha=0.7, edgecolor='black')  # Warna ungu
plt.title('Purchase Price Distribution', fontsize=14)
plt.xlabel('Purchase Price ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


plt.subplot(2, 2, 2)
plt.hist(df['Current Price'], bins=15, color='#E91E63', alpha=0.7, edgecolor='black')  # Warna pink
plt.title('Current Price Distribution', fontsize=14)
plt.xlabel('Current Price ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


plt.subplot(2, 2, 3)
plt.hist(df['Gain/Loss'], bins=8, color='#FF5722', alpha=0.7, edgecolor='black')  # Warna merah
plt.title('Gain/Loss Distribution', fontsize=14)
plt.xlabel('Gain/Loss ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



data = {
    'Stock Name': ['Company A', 'Company B', 'Company C', 'Company D', 'Company E'],
    'Symbol': ['A', 'B', 'C', 'D', 'E'],
    'Dividend per Share': [3.5, 2.8, 4.2, 5.1, 3.9]
}
df = pd.DataFrame(data)


top_dividend_stocks = df.nlargest(5, 'Dividend per Share')[['Stock Name', 'Symbol', 'Dividend per Share']]


plt.figure(figsize=(10, 6))
bars = plt.bar(top_dividend_stocks['Symbol'], top_dividend_stocks['Dividend per Share'], color=np.random.rand(5, 3))


plt.xlabel('Symbol')
plt.ylabel('Dividend per Share')
plt.title('Top 5 Stocks with Highest Dividend per Share')


for bar, div in zip(bars, top_dividend_stocks['Dividend per Share']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, f'{div}', ha='center', va='bottom')


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df = pd.read_csv("stock.csv")

data = {
    'Stock Name': ['Company A', 'Company B', 'Company C', 'Company D', 'Company E'],
    'Symbol': ['A', 'B', 'C', 'D', 'E'],
    'Market Value': [150000000, 180000000, 200000000, 220000000, 250000000]
}
df = pd.DataFrame(data)

top_market_value_stocks = df.nlargest(5, 'Market Value')[['Stock Name', 'Symbol', 'Market Value']]

plt.figure(figsize=(10, 6))

colors = np.random.rand(5, 3)

for i, (stock_name, market_value) in enumerate(zip(top_market_value_stocks['Stock Name'], top_market_value_stocks['Market Value'])):
    plt.plot([i, i, i], [0, market_value, 0], marker='^', markersize=12, color=colors[i], label=stock_name)

plt.xticks(range(len(top_market_value_stocks)), top_market_value_stocks['Symbol'])
plt.xlabel('Symbol')
plt.ylabel('Market Value')
plt.title('Top 5 Stocks with Highest Market Value')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Stock Name': ['Stock A', 'Stock B', 'Stock C', 'Stock D', 'Stock E', 'Stock F', 'Stock G'],
    'Annual Yield': [5.2, 4.8, 4.5, 4.2, 3.9, 3.6, 3.3]
}

df = pd.DataFrame(data)

sorted_data = df.sort_values('Annual Yield', ascending=False)

plt.figure(figsize=(10, 6))

plt.scatter(sorted_data.index, sorted_data['Annual Yield'], s=100, c='blue', marker='^')

for i, txt in enumerate(sorted_data['Stock Name']):
    plt.annotate(txt, (sorted_data.index[i], sorted_data['Annual Yield'][i]), xytext=(10,-10), textcoords='offset points')

plt.title('Top 7 Stocks with Highest Annual Yield', fontsize=14, weight='bold')
plt.xlabel('Stock Index', fontsize=12, weight='bold')
plt.ylabel('Annual Yield (%)', fontsize=12, weight='bold')

plt.xlim(-1, len(sorted_data))

plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

```
